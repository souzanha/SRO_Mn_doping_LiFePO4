```python
path='/home/hammadi/teoroo-folder/tetralith/make_db_for_ccs/MC/224/3-Mn2+_wtCP_LiTcMn_Mnfix/clusters_JK/'

LiFe=pd.read_csv(path+'LiFe_linked.dat',sep="\s+", usecols=[0,1],names=['temperature','fraction'])
LiMn=pd.read_csv(path+'LiMn_linked.dat',sep="\s+", usecols=[0,1],names=['temperature','fraction'])
LiFeMn=pd.read_csv(path+'LiFeMn_linked.dat',sep="\s+", usecols=[0,1],names=['temperature','fraction'])
```

```python
HS_l=1.0540872630243934
HS=2.2368604094358284
```

```python
import matplotlib.font_manager

fig, ax = plt.subplots(1, 1, figsize=(8, 6))

plt.rcParams.update({'font.size': 24})
matplotlib.font_manager.FontProperties(fname='Times New Roman')
plt.rcParams["font.family"] = "Times New Roman"

#ax.plot([0,1020],[HS,HS],":",color='black', label="Hard spheres",linestyle='solid',ms=12,mec='black')
ax.plot([0,1020],[HS_l,HS_l],":",color='black', label="Linked hard spheres",linestyle='dotted',ms=12,mec='black')
ax.plot(Li_linked['temperature'][1:],Li_linked['fraction'][1:],marker='s',color='cornflowerblue',mfc='white',linestyle='dashed',ms=12,mec='black')



ax.plot(LiFe['temperature'],LiFe['fraction'],label='Li-Fe(II)-Li',marker='o',
        color='pink',linestyle='solid',ms=12,mec='black')

ax.plot(LiMn['temperature'],LiMn['fraction'],label='Li-Mn(II)-Li',marker='^',
        color='orange',linestyle='solid',ms=12,mec='black')

ax.plot(LiFeMn['temperature'],LiFeMn['fraction'],label='Li-TM(II)-Li',marker='s',
        color='cornflowerblue',linestyle='solid',ms=12,mec='black')


ax.tick_params(direction='in', which='both',length=10,color='black')

ax.set_xticks([200,300,400,500,600,700,800,900,1000])

ax.set_xlim([180,1020])
ax.set_ylim([0,4])

legend = plt.legend(frameon = 1,fontsize=20)
frame = legend.get_frame()
frame.set_edgecolor('black')

ax.set_xlabel('Temperature [K]')
              
ax.set_ylabel('Average cluster size')#, color='white')

fig.tight_layout()

plt.savefig('Li-Li_cluster_analysis.pdf',dpi=300,bbox_inches='tight',transparent=True)
```
